```python
# Make previous operations a modular function
def data_clean(df):
    """Takes in a dataframe, removes raw values and any SSD devices, prints class baselines
    and returns a dataframe with only HDD."""
    df = df.copy()
    
    # Convert 'date' to datetime
    df['date'] = pd.to_datetime(df['date'], infer_datetime_format=True)
    # Use regular expression to drop columns with _raw data
    df_norm = df[df.columns.drop(list(df.filter(regex='raw')))]
    # Identify any SSDs, save indices to list
    df_smart173 = df_norm[df_norm['smart_173_normalized'] == 100.0]
    index_drop = df_smart173.index.values.tolist()
    # Remove observations from dataframe
    df_hdd= df_norm.drop(index=index_drop)

    # Print class baselines
    print('Raw and normalized counts for Failure class target')
    print(df_hdd['failure'].value_counts())
    print(df_hdd['failure'].value_counts(normalize=True))
    # Print Shape
    print("\nShape of the DataFrame:\n", df_hdd.shape)
    
    
    return df_hdd
```

```python
# Import Concatenated Data
df_concat = pd.read_csv('/home/tyler/Documents/Data Subset/combined_csv.csv')
print(df_concat.shape)
df_concat.head()
```

```python
df_concat = data_clean(df_concat)
df_concat.head()
```
